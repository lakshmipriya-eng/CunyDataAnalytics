```python
def winner():
  p1 = input("p1: please enter rock, paper, scissors")
  p2 = input("p2: please enter rock, paper,scissors")

  if p1 == p2:
    print("tie")
  elif p1 == "rock" and p2 == "scissors":
    print("p1 won")
  elif p1 == "rock" and p2 == "paper":
    print("p2 won")
  elif p1 == "paper" and p2 == "rock":
    print("p1 won")
  elif p2 == "rock" and p1 == "scissors":
    print("p2 won")
  elif p2 == "rock" and p1 == "paper":
    print("p1 won")
  elif p2 == "paper" and p1 == "rock":
    print("p2 won")
  else:
    print("congratulations")

winner()
```
